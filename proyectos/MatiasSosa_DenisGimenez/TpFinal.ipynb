```python
import pandas as pd
import gzip
import json

def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield json.loads(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')
```

## get_list_also_view_buy
- Se utiliza para obtener un merge de la lista de also_buy y also_view, sin repetir los elementos
- Esto se utiliza para el calculo de la similaridad de jaccard

```python
import math
def get_list_also_view_buy(df_metadata, asin_prod):
    '''
    Retorna una sola lista con los productos de also_view y also_buy para un producto específico.

    Args:
        df_metadata (pandas.DataFrame): El DataFrame que contiene los datos.
        asin_prod (str): El ASIN del producto del cual se desea obtener la lista de productos también vistos y comprados.

    Returns:
        list: Una lista que contiene los elementos únicos de las listas de also_view y also_buy del producto especificado.
    
    Raises:
        SystemExit: Si existen múltiples repeticiones del mismo producto en el DataFrame.
    
    Example:
        >>> import pandas as pd
        >>> data = {'asin': ['A', 'B', 'A'],
        ...         'also_view': [['X', 'Y'], ['Y', 'Z'], ['X', 'Z']],
        ...         'also_buy': [['P', 'Q'], ['Q', 'R'], ['P', 'R']]}
        >>> df_metadata = pd.DataFrame(data)
        >>> get_list_also_view_buy(df_metadata, 'A')
        ['X', 'Y', 'P', 'Q']

    '''
    df_metadata_prod = df_metadata[df_metadata['asin'] == asin_prod]
    if len(df_metadata_prod) == 1: 
        lista_also_view_prod = df_metadata_prod['also_view'].tolist()[0]
        # Si la lista contiene un valor NaN entonces se asigna una lista vacia
        if not isinstance(lista_also_view_prod, list):
            lista_also_view_prod = []

        # Si la lista contiene un valor NaN entonces se asigna una lista vacia
        lista_also_buy_prod = df_metadata_prod['also_buy'].tolist()[0]
        if not isinstance(lista_also_buy_prod, list):
            lista_also_buy_prod = []

        lista_elementos_prod = list(set(lista_also_view_prod) | set(lista_also_buy_prod))

        return lista_elementos_prod
    elif len(df_metadata_prod) > 1:
        lista_elementos_prod = []
        for i in range(len(df_metadata_prod)):
            lista_also_view_prod = df_metadata_prod['also_view'].tolist()[i]
            lista_also_buy_prod = df_metadata_prod['also_buy'].tolist()[i]
            lista_elementos_prod = list(set(lista_also_view_prod) | set(lista_also_buy_prod) | set(lista_elementos_prod))
        return lista_elementos_prod
    else:
        print('No existe el producto en el DataFrame.')
```

## Similaridad de jaccard(Diferente a la implementacion de cosine del pdf)*
- Se opto por este calculo ya que es mas eficiente que la representacion vectorial de los productos
La similitud de Jaccard es una métrica utilizada para calcular la similitud entre dos conjuntos, en este caso, se aplica a conjuntos de productos "also_views" y "also_buy" de dos productos. La fórmula de la similitud de Jaccard se define de la siguiente manera:

$$ J(A, B) = \frac{|A \cap B|}{|A \cup B|} $$

Donde:
- $A$ es el conjunto de elementos relacionados con el producto 1 (also_view y also_buy).
- $B$ es el conjunto de elementos relacionados con el producto 2 (also_view y also_buy).
- $|A \cap B|$ es el tamaño de la intersección de los conjuntos $A$ y $B$, es decir, cuántos elementos tienen en común.
- $|A \cup B|$ es el tamaño de la unión de los conjuntos $A$ y $B$, es decir, cuántos elementos tienen en total entre ambos conjuntos.

El código implementa la similitud de Jaccard entre dos productos utilizando los conjuntos de productos "also_views" y "also_buy" en un DataFrame de metadatos de productos.

1. Se definió la función $\texttt{jaccard\_similarity}$ que toma tres argumentos:
   - $\texttt{df\_metadata}$: El DataFrame que contiene la metadata de los productos.
   - $\texttt{asin\_prod1}$: El ASIN del primer producto.
   - $\texttt{asin\_prod2}$: El ASIN del segundo producto.

2. La función llama a $\texttt{get\_list\_also\_view\_buy}$ para obtener las listas de productos "also_views" y "also_buy" para los dos productos dados.

3. Luego, encuentra la intersección de las dos listas usando $\texttt{set(prod1\_lista\_also\_buy\_view) \& set(prod2\_lista\_also\_buy\_view)}$ para obtener los elementos en común entre las dos listas.

4. Calcula la unión de las dos listas utilizando $\texttt{set(prod1\_lista\_also\_buy\_view) | set(prod2\_lista\_also\_buy\_view)}$ para obtener todos los elementos únicos de ambas listas.

5. Finalmente, calcula la similitud de Jaccard dividiendo el tamaño de la intersección entre el tamaño de la unión. Si la unión es vacía, devuelve 0.


```python
def jaccard_similarity(df_metadata, asin_prod1, asin_prod2):
    '''
    Calcula la similitud de Jaccard entre dos productos basándose en los productos también vistos y comprados.

    Args:
        df_metadata (pandas.DataFrame): El DataFrame que contiene la metadata de los productos.
        asin_prod1 (str): El ASIN del primer producto.
        asin_prod2 (str): El ASIN del segundo producto.

    Returns:
        float: El coeficiente de similitud de Jaccard entre los dos productos basado en los productos también vistos y comprados.
    
    Example:
        >>> import pandas as pd
        >>> data = {'asin': ['A', 'B', 'A'],
        ...         'also_view': [['X', 'Y'], ['Y', 'Z'], ['X', 'Z']],
        ...         'also_buy': [['P', 'Q'], ['Q', 'R'], ['P', 'R']]}
        >>> df_metadata = pd.DataFrame(data)
        >>> jaccard_similarity(df_metadata, 'A', 'B')
        0.25

    '''
    prod1_lista_also_buy_view = get_list_also_view_buy(df_metadata, asin_prod1)
    prod2_lista_also_buy_view = get_list_also_view_buy(df_metadata, asin_prod2)

    interseccion_lista_also_buy_view = list(set(prod1_lista_also_buy_view) & set(prod2_lista_also_buy_view))

    union_lista_also_buy_view = list(set(prod1_lista_also_buy_view) | set(prod2_lista_also_buy_view))
    if len(union_lista_also_buy_view) != 0:
        return len(interseccion_lista_also_buy_view)/len(union_lista_also_buy_view)
    else:
        return 0
```

## get_prod_overall_rating
- Simplemente calcula el promedio de overall de todas las reviews de un asin de un producto especifico

```python
def get_prod_overall_rating(df_reviews, asin_prod):
    """
    Calcula el promedio de calificaciones generales (overall) de un producto específico en un DataFrame de reseñas.

    Args:
        df_reviews (DataFrame): El DataFrame que contiene las reseñas de productos.
        asin_prod (str): El código ASIN del producto del cual se desea calcular el promedio de calificaciones.

    Returns:
        float: El promedio de calificaciones generales (overall) del producto especificado.

    Example:
        Si tenemos un DataFrame 'df_reviews' con reseñas de productos y queremos calcular el promedio de calificaciones
        generales para el producto con ASIN 'B001GXRQW0', podemos hacerlo de la siguiente manera:

        >>> rating_score_prod = get_prod_overall_rating(df_reviews, 'B001GXRQW0')
        >>> print(rating_score_prod)
        4.09
    """
    return df_reviews[df_reviews['asin'] == asin_prod]['overall'].mean()
```

Si similarity_score es bajo y rating_score es medio, entonces recomendacion es baja.

3. Regla 3:<br>
Si similarity_score es bajo y rating_score es alto, entonces recomendacion es media.

4. Regla 4:<br>
Si similarity_score es medio y rating_score es bajo, entonces recomendacion es baja.

5. Regla 5:<br>
Si similarity_score es medio y rating_score es medio, entonces recomendacion es media.

6. Regla 6:<br>
Si similarity_score es medio y rating_score es alto, entonces recomendacion es alta.

7. Regla 7:<br>
Si similarity_score es alto y rating_score es bajo, entonces recomendacion es media.

8. Regla 8:<br>
Si similarity_score es alto y rating_score es medio, entonces recomendacion es alta.

9. Regla 9:<br>
Si similarity_score es alto y rating_score es alto, entonces recomendacion es alta.


- En el codigo las posiciones de las respectivas listas son [0] bajo, [1] medio, [2], alto
- resultado: 1(bajo), 2(medio), 3(alto) este es para el valor final que retorna

```python
def get_rules_evaluation(df_metadata, df_reviews, prod1_asin, prod2_asin, similarity_score_prod2, rating_score_prod2, antecedentes_similarity, antecedentes_rating_score, consecuentes_recomendacion):
    """
    Evalúa la recomendación de un producto en función de reglas difusas.

    Esta función calcula la recomendación de un producto basándose en reglas difusas definidas previamente. 
    Utiliza el similarity_score y el product_rating para determinar el nivel de recomendación.

    Args:
        df_metadata (pandas.DataFrame): El DataFrame que contiene la metadata de los productos.
        df_reviews (pandas.DataFrame): El DataFrame que contiene las reseñas de los productos.
        prod1_asin (str): El ASIN del primer producto.
        prod2_asin (str): El ASIN del segundo producto.
        similarity_score_prod2 (float): El valor de similitud entre los productos.
        rating_score_prod2 (float): La puntuación de calificación del segundo producto.
        antecedentes_similarity (list): Una lista de tres valores de similitud utilizados como antecedentes en las reglas difusas.
        antecedentes_rating_score (list): Una lista de tres valores de puntuación de calificación utilizados como antecedentes en las reglas difusas.
        consecuentes_recomendacion (list): Una lista de tres valores de recomendación que representan los consecuentes en las reglas difusas.

    Returns:
        int: Un valor que representa el nivel de recomendación del producto:
             - 1: No recomendado.
             - 2: Medianamente recomendado.
             - 3: Recomendado.
             - 0: Valor no válido.

    Example:
        >>> import pandas as pd
        >>> data = {'asin': ['A', 'B', 'A'],
        ...         'also_view': [['X', 'Y'], ['Y', 'Z'], ['X', 'Z']],
        ...         'also_buy': [['P', 'Q'], ['Q', 'R'], ['P', 'R']]}
        >>> df_metadata = pd.DataFrame(data)
        >>> df_reviews = pd.DataFrame({'asin': ['A', 'B'], 'overall': [4.5, 3.0]})
        >>> antecedentes_similarity = [[0.4, 0.6, 0.8]]
        >>> antecedentes_rating_score = [[2.0, 3.0, 4.5]]
        >>> consecuentes_recomendacion = [[1, 2, 3]]
        >>> get_rules_evaluation(df_metadata, df_reviews, 'A', 'B', 0.7, 3.5, antecedentes_similarity, antecedentes_rating_score, consecuentes_recomendacion)
        2
    """
    similarity_score_prod2 = jaccard_similarity(df_metadata, prod1_asin, prod2_asin)
    rating_score_prod2 = get_prod_overall_rating(df_reviews, prod2_asin)

    # Regla 1
    b_prima, y0 = algoritmo(similarity_score_prod2, rating_score_prod2, antecedentes_similarity[0], antecedentes_rating_score[0], consecuentes_recomendacion[0])
    recomendacion_baja = max(y0, 0)

    # Regla 2
    b_prima, y0 = algoritmo(similarity_score_prod2, rating_score_prod2, antecedentes_similarity[0], antecedentes_rating_score[1], consecuentes_recomendacion[0])
    recomendacion_baja = max(y0, recomendacion_baja)

    # Regla 3
    b_prima, y0 = algoritmo(similarity_score_prod2, rating_score_prod2, antecedentes_similarity[0], antecedentes_rating_score[2], consecuentes_recomendacion[1])
    recomendacion_media = max(y0, 0)

    # Regla 4
    b_prima, y0 = algoritmo(similarity_score_prod2, rating_score_prod2, antecedentes_similarity[1], antecedentes_rating_score[0], consecuentes_recomendacion[0])
    recomendacion_baja = max(y0, recomendacion_baja)
    
    # Regla 5
    b_prima, y0 = algoritmo(similarity_score_prod2, rating_score_prod2, antecedentes_similarity[1], antecedentes_rating_score[1], consecuentes_recomendacion[1])
    recomendacion_media = max(y0, recomendacion_media)

    # Regla 6
    b_prima, y0 = algoritmo(similarity_score_prod2, rating_score_prod2, antecedentes_similarity[1], antecedentes_rating_score[2], consecuentes_recomendacion[2])
    recomendacion_alta = max(y0, 0)
    
    # Regla 7
    b_prima, y0 = algoritmo(similarity_score_prod2, rating_score_prod2, antecedentes_similarity[2], antecedentes_rating_score[0], consecuentes_recomendacion[1])
    recomendacion_media = max(y0, recomendacion_media)

    # Regla 8
    b_prima, y0 = algoritmo(similarity_score_prod2, rating_score_prod2, antecedentes_similarity[2], antecedentes_rating_score[1], consecuentes_recomendacion[2])
    recomendacion_alta = max(y0, recomendacion_alta)

    # Regla 9
    b_prima, y0 = algoritmo(similarity_score_prod2, rating_score_prod2, antecedentes_similarity[2], antecedentes_rating_score[2], consecuentes_recomendacion[2])
    recomendacion_alta = max(y0, recomendacion_alta)

    if recomendacion_baja >= recomendacion_media and recomendacion_baja >= recomendacion_alta:
        return 1
    elif recomendacion_media >= recomendacion_baja and recomendacion_media >= recomendacion_alta: 
        return 2
    elif recomendacion_alta >= recomendacion_baja and recomendacion_alta >= recomendacion_media: 
        return 3
    else:
        return 0
```

## Ejecucion final del codigo
- Se elige un asin code de un producto especifico y se compara cada producto del dataframe
- Esta implementacion es multihilo, pero la diferencia es muy poca en tiempo de ejecucion
- Retorna un diccionario {resultado: valor_recomendacion, asin: producto_a_recomendar}

```python
import concurrent.futures

def worker(prod, df_metadata, df_reviews, prod1_asin, antecedentes_similarity, antecedentes_rating_score, consecuentes_recomendacion):
    similarity_score_prod = jaccard_similarity(df_metadata, prod1_asin, prod.asin)
    rating_score_prod = get_prod_overall_rating(df_reviews, prod.asin)
    resultado = get_rules_evaluation(df_metadata, df_reviews, prod1_asin, prod.asin, similarity_score_prod, rating_score_prod, antecedentes_similarity, antecedentes_rating_score, consecuentes_recomendacion)
    
    return {"resultado": resultado, "asin": prod.asin, 'similarity':similarity_score_prod, 'rating':rating_score_prod}

def process_with_multithreading(df_metadata, df_reviews, prod1_asin, antecedentes_similarity, antecedentes_rating_score, consecuentes_recomendacion, num_threads):
    results_list = []
    
    with concurrent.futures.ThreadPoolExecutor(max_workers=num_threads) as executor:
        futures = [executor.submit(worker, prod, df_metadata, df_reviews, prod1_asin, antecedentes_similarity, antecedentes_rating_score, consecuentes_recomendacion) for prod in df_metadata.itertuples() if prod.asin != prod1_asin] 
        for future in concurrent.futures.as_completed(futures):
            result_dict = future.result()
            print(result_dict)
            if result_dict["resultado"] >= 3:
                results_list.append(result_dict)

    return results_list

# Uso de la función:
num_threads = 16  # Por ejemplo, 4 hilos
prod1_asin = 'B001H53QE4'
if len(df_metadata[df_metadata['asin'] == prod1_asin])>0:
    results = process_with_multithreading(df_metadata, df_reviews, prod1_asin, antecedentes_similarity, antecedentes_rating_score, consecuentes_recomendacion, num_threads)
else:
    print('No existe el producto en el DataFrame.')
```
